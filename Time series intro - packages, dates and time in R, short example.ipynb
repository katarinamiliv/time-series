#### 1) Small exercise: using 'lubridate' on a data frame

```r
library('lubridate')

# creating a df that contains date and time data (vector a for date, and vector b for time), as well as random measurement (f)
a = c("1998,11,11", "1983/01/23", "1982:09:04", "1945-05-09", 19821224, "1974.12.03", 19871210)
a = ymd(a, tz = "CET"); # CEST for summertime
a
```

```r
b = c("22 4 5", "04;09;45", "11:9:56", "23,15,12", "14 16 34", "8 8 23", "21 16 14")
b = hms(b); 
b
```

```r
# creating a random measurement column rounded on two decimals
f = rnorm(7,10); 
f = round(f, digits = 2); 
f
```

```r
# putting date, time and random measurement in one single data frame
date_time_measurement = cbind.data.frame(date = a, time = b, measurement = f)
date_time_measurement
```

# Example: IT Store Staff Allocation

### The data contains the number of customers who entered the store from Monday to Saturday in two shifts, during 12 weeks (the total of 144 data points).

```r
# import data
myts <- read.csv('ITstore-bidaily.csv', header = FALSE, sep = ';')
```

```r
# check how it looks
head(myts)
```

```r
# convert data to a time series object
# start is the first available value of the first cycle; frequency is the nr of observations per cycle 
# here we had 12 observations per week (Mon to Sat, first and second shift)
mycounts <- ts(myts$V2, start = 1, frequency = 12) 

# checking the type of object and its class
typeof(mycounts) # 'integer'
class(mycounts) # class is 'ts'
```

```r
# visualization: we can detect seasonality along 12 weeks - highest customer counts appear to be at the end of each week
plot(mycounts, main = 'F of customers in two shifts from Mon to Sat during 12w', ylab = 'Customer Counts', xlab = 'Weeks', ylim = c(0,700))
axis(side = 1, at = seq(1, 12, 1))
```

```r
library(forecast) # for month plot and season plot

# month plot compares the single time units (shifts for each of six days in the week) within the seasonal unit (one week) 
# horizontal lines present average number of customers in among the seasons for each time unit
# here we see that highest customer counts were on Saturday evening, Saturday afternoon, Friday evening, and Monday evening)
monthplot(mycounts, labels = 1:12, xlab = 'Bidaily Units', ylab = 'Customer Counts', ylim = c(0,700))
```

```r
# season plot compares the seasonal units (cycles, i.e. weeks) to one another
# here we see that the similar pattern appeared each week
seasonplot(mycounts, season.labels = T, xlab = 'Bidaily Units', ylab = 'Customer Counts', ylim = c(0,700))
```
