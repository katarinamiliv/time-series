# Time series formatting

```r
# initializing a random walk of 450 numbers
set.seed(41) # for a reproducible result
x = cumsum(rnorm(n = 450)) # rnorm for normal distribution, runif for uniform distribution
typeof(x)
head(x)
```

```r
# creating a time series object: starting in November 1914, time unit is a month, and seasonal unit is a year
time_series <- ts(x, start = c(1914,11), frequency = 12)
typeof(time_series)
```

# Visualization
## Visualizing the Nottingham temperature data

### check 'shiny' https://shiny.rstudio.com/

```r
# visualization with standard R base plot
plot(nottem)
```

```r
# observed = trend + seasonal + random noise
plot(decompose(nottem))
```

```r
# different visualization with library 'lattice'
library('lattice')
xyplot.ts(nottem)
```

```r
# visualization with 'ggplot2'
library('forecast')
library('ggplot2')
autoplot(nottem) + ggtitle('Autoplot of Nottingham temperature data')
```

```r
# month plot with 'ggplot2'
ggmonthplot(nottem) # How the temperature of each month develops over years + the mean horizontal line
```

```r
# season plot with 'ggplot2'
ggseasonplot(nottem) # there is a pattern - summers have highest temperatures
```

## Seasonal plot of data set AirPassengers

```r
seasonplot(AirPassengers, 
           xlab = "",                    # deleting x labels
           col = c("green", "violet"),    # two colors of lines
           year.labels = T,              # labels for each year to appear on the graph
           labelgap = 0.35,              # space between the lines and the labels
           type = "l",                   # 'l' for line chart, 'p' for points, 's' for steps (often in survival analysis plots)
           bty = "7", cex = 0.75,        # bty for box lines, cex for size of the plot's features
           main = "Seasonal plot of dataset AirPassengers")
```
